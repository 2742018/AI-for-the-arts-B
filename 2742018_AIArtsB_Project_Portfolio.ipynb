# Access Through Touch: Disability, Reading, and Accessibility Through the Raised-Type New Testament

## Introduction

This project examines a tactile New Testament printed in high relief for blind readers, held in the University of Glasgow Special Collections under the shelfmarks Sp Coll RF 256, vol. 1 and Sp Coll RF 257, vol. 2. The University of Glasgow Library catalogue identifies the object as Bible. New Testament. English. Authorized. 1850, published in New York by the American Bible Society in 1850. 

The object is significant because it challenges a common assumption about reading: that reading is mainly visual. Unlike an ordinary printed book, this New Testament was designed so that its letters could be read through touch. Its raised letterforms, spacing, and material surface invite us to understand reading as an embodied activity shaped by design, technology, and access.

Within this scope, the project introduces the tactile New Testament through a public-facing digital mini-exhibit titled Access Through Touch. While the digital platform cannot replicate the tactile experience, the project employs photographs, relief-enhanced visual aids, plain-language interpretation, and access-description writing. These tools aim to help a general audience understand the New Testament as a historical accessibility technology.

The central argument of the project is that accessibility constitutes an integral aspect of cultural heritage. Accessibility should not be regarded solely as a contemporary support feature appended after collections have been interpreted. This object demonstrates that access has a historical dimension. It illustrates the intersection of religious knowledge, literacy, disability, and material design in the nineteenth century. By highlighting the tactile qualities of the page, the project invites viewers to reflect on for whom books are created, who is envisioned as a reader, and what may be lost when tactile heritage is digitised exclusively as flat images.

The project incorporates human–AI collaboration within its creative and interpretive processes. Generative AI assists in drafting alternative labels, access-description text, and explanatory language. Python is used to generate visual aids that enhance the visibility of surface details in a digital context. However, AI is not regarded as an authoritative source. Human contributors retain responsibility for object selection, fragment choice, factual verification, language editing, ethical risk assessment, and decisions regarding public presentation.

## Human–AI Collaboration

AI serves as a collaborator in the public interpretation process, not as an independent curator or authority. The human role begins with selecting the object and deciding to frame it around accessibility. This is an interpretive decision. The generative AI role primarily occurs during the drafting stage. For example, a text model can produce several versions of a plain-language exhibit label. Some versions may be useful because they offer clearer phrasing or alternative ways into the object. Other versions may be too vague, too dramatic, or overly sentimental about blindness. The human author therefore revises the text to make it more accurate, respectful, and useful for a general audience.

Alongside generative text models, Python code also plays a vital, but distinct, role in the workflow. Unlike generative AI, this code does not create new content from training data. Instead, it transparently processes the project’s own photographs into cropped views and relief-enhanced visual aids, helping viewers notice surface details that standard photographs might overlook. This computational aspect complements the interpretive work described above.

The collaboration can therefore be understood as a workflow:

| Stage | Human contribution | AI/computational contribution | Public value |
|---|---|---|---|
| Object framing | Selects the object and defines accessibility as the central theme | No AI contribution at this stage | Gives the exhibit a clear social and cultural purpose |

This means the project's creative value does not come solely from AI. In this project, AI expands the range of possible explanations, but the human author remains responsible for meaning, evidence, access, and integrity.

## How Generative AI Works in This Project

Generative AI refers to artificial intelligence systems that produce new content based on patterns learned from training data. In this project, generative AI is used mainly for text generation: drafting possible exhibit labels, access descriptions, and explanatory wording for a general audience.

To understand how this works, consider the text-based AI tool through the lens of a basic machine-learning workflow:

| Machine-learning stage | How it appears in this project |
|---|---|
| Training data | The model has previously been trained on large collections of text. The exact training data is not fully visible to the user, which creates limits around transparency and source attribution. |
| Algorithm/model | The model uses a transformer-based architecture. It generates text by predicting likely sequences of words in response to a prompt. |
| New input | The human author provides prompts describing the object, audience, desired tone, and ethical constraints. |
| Output | The model produces draft labels, access descriptions, or explanations. |
| Human verification | The draft is checked against the object, the project aims, and accessibility/ethical requirements before being used. |

Transformer models are useful for this project because they can quickly generate fluent alternative phrasings. This supports public interpretation by helping test different ways of explaining the same object. However, fluency does not guarantee truth. A model may produce confident but unsupported claims, especially when asked about a specific archival object. For this reason, AI-generated text is treated as draft material rather than archival evidence.

In addition to generative AI, the project also employs Python image processing, which operates with a different approach. Unlike generative AI, the Python code does not create a new imagined object or reconstruct missing features. Instead, it applies explicit computational transformations to existing photographs: cropping, grayscale conversion, contrast enhancement, sharpening, and edge detection. These steps help sighted online viewers notice surface variation, but the resulting images are visual aids rather than scientific measurements of texture or substitutes for touch.

The project, therefore, distinguishes between four types of material:

| Material type | Role in the project |
|---|---|
| **Archival evidence** | The original object and photographs of it. |
| **Computational visual aids** | Processed images that help viewers notice surface features in a flat photograph. |
| **AI-assisted interpretation** | Draft text generated by AI and edited by the human author. |
| **Human judgement** | The final decision-making process that checks meaning, tone, ethics, accuracy, and accessibility. |

Recognising these different tool functions is crucial, as the project brings together several approaches, each carrying a different level of authority. The photographs are closest to evidence; the relief-enhanced images and AI-assisted labels are interpretive layers built around that evidence.

### Step 3: Creating a contact sheet

The contact sheet provides the visual entry point to the mini-exhibit. It gives viewers a quick overview of the selected photographs before the notebook moves into closer interpretation of individual fragments.

It also supports transparency and reproducibility. Because the contact sheet is generated directly from the images in the repository, the visual overview can be recreated whenever the notebook is rerun.

```python
# Building a contact sheet as the visual entry point to the exhibit.
# The contact sheet is generated from the repository images so that the notebook remains reproducible.

thumbs = []

for path in image_paths:
    im = Image.open(path).convert("RGB")
    im.thumbnail((220, 220))
    thumbs.append((path.name, im.copy()))

cols = 2
cell_w, cell_h = 520, 285
rows = int(np.ceil(len(thumbs) / cols))

sheet = Image.new("RGB", (cols * cell_w, rows * cell_h), "white")
draw = ImageDraw.Draw(sheet)

for idx, (name, im) in enumerate(thumbs):
    x = (idx % cols) * cell_w + 20
    y = (idx // cols) * cell_h + 20
    sheet.paste(im, (x, y))
    draw.text((x + 240, y + 70), name, fill="black")

contact_path = CONTACT_DIR / "selected_images_contact_sheet.jpg"
sheet.save(contact_path, quality=90)

display(IPImage(filename=str(contact_path)))
```

The contact sheet acts as the visual entry point to the mini-exhibit. It gives viewers an overview of the selected photographs before the notebook moves into closer interpretation of individual fragments.

For the project workflow, the contact sheet also supports reproducibility. It is generated directly from the image files in the `images/` folder rather than manually inserted as a separate design element.

### Step 5: Creating relief-enhanced visual aids

This stage uses Python to create **relief-enhanced visual aids** from the object photographs. They are described as visual aids rather than scans or measurements because the code does not measure the physical height of the raised letters. Instead, it transforms the photograph so that contrast, edges, and surface variation are easier to notice.

This supports the project’s public aim. Online visitors cannot touch the page, so the processed image helps draw attention to the object’s tactile qualities. However, it does not reproduce tactile reading. The relief-enhanced image remains an interpretive aid, not a replacement for the physical object.

The code below defines two functions: one for cropping selected areas of the photograph and one for creating the relief-enhanced visual aid.

```python
# Functions for cropping and creating relief-enhanced visual aids.

def crop_by_percentage(image, crop_pct):
    """
    Crop an image using percentage coordinates.

    crop_pct should be a tuple:
    (left, top, right, bottom)

    Each value is between 0 and 1.
    For example, (0.10, 0.10, 0.90, 0.90) keeps the central part of the image.
    """
    width, height = image.size

    left = int(width * crop_pct[0])
    top = int(height * crop_pct[1])
    right = int(width * crop_pct[2])
    bottom = int(height * crop_pct[3])

    return image.crop((left, top, right, bottom))


def create_relief_visual_aid(image):
    """
    Create an interpretive visual aid that makes surface variation easier to notice.

    This does not measure physical relief. It enhances contrast and edge information
    in the photograph so that sighted online viewers can better notice surface detail.
    """
    # Convert to grayscale so that tonal differences are easier to process.
    gray = ImageOps.grayscale(image)

    # Increase contrast to make surface variation more noticeable.
    contrast = ImageEnhance.Contrast(gray).enhance(1.8)

    # Sharpen image details.
    sharp = ImageEnhance.Sharpness(contrast).enhance(2.0)

    # Detect edges and increase their contrast.
    edges = sharp.filter(ImageFilter.FIND_EDGES)
    edges = ImageEnhance.Contrast(edges).enhance(1.7)

    # Blend the sharpened image with the edge-detected image.
    blended = Image.blend(sharp.convert("RGB"), edges.convert("RGB"), 0.35)

    return blended


print("Cropping and relief-enhancement functions are ready.")
```

The next code cell applies these functions to three selected fragments. These fragments are chosen because they help the audience understand the page as a tactile reading surface, not only as a flat image.

```python
# Generate cropped images and relief-enhanced visual aids.

# This safety function allows the notebook to run even if there are fewer than three images.
# Python counts from 0, so choose_image(0) means the first image.
if "choose_image" not in globals():
    def choose_image(index):
        return image_paths[min(index, len(image_paths) - 1)]


fragment_plan = [
    {
        "fragment": "Fragment 1",
        "source": choose_image(4),
        "crop_pct": (0.05, 0.05, 0.95, 0.75)
    },
    {
        "fragment": "Fragment 2",
        "source": choose_image(6),
        "crop_pct": (0.10, 0.10, 0.90, 0.80)
    },
    {
        "fragment": "Fragment 3",
        "source": choose_image(2),
        "crop_pct": (0.05, 0.15, 0.95, 0.90)
    }
]


fragment_records = []

for item in fragment_plan:
    source_path = item["source"]
    original = Image.open(source_path).convert("RGB")

    # Crop the selected part of the image.
    cropped = crop_by_percentage(original, item["crop_pct"])
    cropped.thumbnail((900, 900))

    # Create the relief-enhanced visual aid.
    relief = create_relief_visual_aid(cropped)

    # Create safe filenames.
    fragment_id = item["fragment"].lower().replace(" ", "_")

    crop_path = PROCESSED_DIR / f"{fragment_id}_crop.jpg"
    relief_path = PROCESSED_DIR / f"{fragment_id}_relief_visual_aid.jpg"

    # Save outputs.
    cropped.save(crop_path, quality=90)
    relief.save(relief_path, quality=90)

    fragment_records.append({
        "fragment": item["fragment"],
        "source image": source_path.name,
        "crop path": str(crop_path.relative_to(PROJECT_ROOT)),
        "relief visual aid path": str(relief_path.relative_to(PROJECT_ROOT))
    })


processed_fragments = pd.DataFrame(fragment_records)
display(processed_fragments)
```

The table above records the generated image outputs. The cropped image provides a closer photographic view, while the relief-enhanced image makes surface variation easier to notice.

This step is part of the exhibit rather than separate technical evidence. The visual aids help the audience understand that the raised forms on the page are functional. They are part of how the book was designed to be read.

### Step 6: Comparing photographic and relief-enhanced views

This step compares each photographic crop with its relief-enhanced version. The comparison is important because it makes clear what the Python image processing contributes to the public interpretation.

The original crop keeps the viewer close to the photographic evidence, while the relief-enhanced version makes tonal changes, edges, and surface variation easier to notice. Placing them side by side helps the audience understand both the value and the limits of the digital process.

Together, the two views support the project’s central argument: the raised-type page is meaningful because of its tactile design, but a digital image can only point towards that material experience. It can help viewers notice the surface, but it cannot reproduce the act of reading by touch.

```python
# Code Cell 6: Create side-by-side comparison images.

def resize_to_height(image, target_height):
    """
    Resize an image to a target height while preserving its proportions.
    """
    ratio = target_height / image.height
    new_width = int(image.width * ratio)
    return image.resize((new_width, target_height))


def make_comparison_image(crop_path, relief_path, title):
    """
    Create a side-by-side comparison image for the mini-exhibit.
    """
    crop = Image.open(crop_path).convert("RGB")
    relief = Image.open(relief_path).convert("RGB")

    target_height = 420

    crop = resize_to_height(crop, target_height)
    relief = resize_to_height(relief, target_height)

    padding = 30
    title_h = 60
    label_h = 40

    width = crop.width + relief.width + padding * 3
    height = target_height + title_h + label_h + padding

    canvas = Image.new("RGB", (width, height), "white")
    draw = ImageDraw.Draw(canvas)

    draw.text((padding, 20), title, fill="black")
    draw.text((padding, title_h), "Photographic crop", fill="black")
    draw.text((crop.width + padding * 2, title_h), "Relief-enhanced visual aid", fill="black")

    canvas.paste(crop, (padding, title_h + label_h))
    canvas.paste(relief, (crop.width + padding * 2, title_h + label_h))

    return canvas


comparison_paths = []

for _, row in processed_fragments.iterrows():
    crop_path = PROJECT_ROOT / row["crop path"]
    relief_path = PROJECT_ROOT / row["relief visual aid path"]

    comparison = make_comparison_image(
        crop_path,
        relief_path,
        f"{row['fragment']}: photograph and interpretive visual aid"
    )

    comparison_path = PROCESSED_DIR / f"{row['fragment'].lower().replace(' ', '_')}_comparison.jpg"
    comparison.save(comparison_path, quality=90)

    comparison_paths.append(comparison_path)

    display(IPImage(filename=str(comparison_path)))
```

The comparison images are the technical centre of the mini-exhibit. They show how a flat photograph can be transformed into a visual aid that supports public interpretation.

The processed image may make surface detail easier to notice, but it should not be confused with the physical surface of the page. This distinction matters because the object’s meaning depends on touch, and touch cannot be fully reproduced in a Jupyter Notebook.

The comparison therefore has two purposes: it helps viewers notice the raised forms more clearly, and it also makes the limits of digital representation visible.

### Step 7: A small technical summary

This step gives a small technical summary of the image transformation. It calculates average brightness and a simple edge-density estimate for each photographic crop and each relief-enhanced visual aid.

These values are not scientific measurements of the raised type. They do not prove where the raised letters are or how high they are. Instead, they act as a descriptive check on the image processing, showing that the relief-enhanced images have changed the image data in a visible and explainable way.

This supports the project’s transparency: the visual aids are not mysterious or manually altered without explanation. They are produced through a clear computational process that can be rerun and inspected.

```python
# Creating a small technical summary.
# This is not a scientific measurement of raised type.
# It only describes simple visual differences between original crops and processed visual aids.

def image_summary(path):
    """
    Return simple brightness and edge-density values for an image.

    Brightness is the average pixel value.
    Edge density is estimated from changes between neighbouring pixels.
    """
    image = Image.open(path).convert("L").resize((400, 400))
    arr = np.asarray(image).astype(float)

    brightness = arr.mean()

    horizontal_change = np.abs(np.diff(arr, axis=1)).mean()
    vertical_change = np.abs(np.diff(arr, axis=0)).mean()
    edge_density = horizontal_change + vertical_change

    return brightness, edge_density


summary_rows = []

for _, row in processed_fragments.iterrows():
    crop_path = PROJECT_ROOT / row["crop path"]
    relief_path = PROJECT_ROOT / row["relief visual aid path"]

    crop_brightness, crop_edges = image_summary(crop_path)
    relief_brightness, relief_edges = image_summary(relief_path)

    summary_rows.append({
        "fragment": row["fragment"],
        "crop average brightness": round(crop_brightness, 2),
        "crop edge-density estimate": round(crop_edges, 2),
        "relief aid average brightness": round(relief_brightness, 2),
        "relief aid edge-density estimate": round(relief_edges, 2)
    })


technical_summary = pd.DataFrame(summary_rows)
display(technical_summary)
```

The edge-density estimate is included carefully. It does not measure the physical relief of the page; it only describes the amount of local visual change within the image.

For the audience, this supports the visual task of noticing surface variation. For the methodology, it shows that the visual aids are generated through a transparent computational process rather than edited without explanation.

This technical step links directly to the project’s public aim: helping viewers understand that a tactile book cannot be fully interpreted through an unprocessed flat image alone. The numbers do not replace interpretation, but they help explain what the image processing is doing.
