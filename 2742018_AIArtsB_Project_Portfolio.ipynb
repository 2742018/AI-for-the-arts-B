# Access Through Touch: Disability, Reading, and Accessibility Through the Raised-Type New Testament

## Introduction

This project explores a tactile New Testament printed in high relief for blind readers, held in the University of Glasgow Special Collections as **Sp Coll RF 256 / RF 257**. The object is significant because it challenges a common assumption about reading: that reading is mainly visual. Unlike an ordinary printed book, this New Testament was designed so that its letters could be read through touch. Its raised letterforms, spacing, and material surface invite us to understand reading as an embodied activity shaped by design, technology, and access.

The project presents the object through a small public-facing digital mini-exhibit titled **Access Through Touch**. The aim is not to reproduce the experience of touching the object, because a digital notebook cannot recreate pressure, texture, or fingertip movement. Instead, the project uses photographs, relief-enhanced visual aids, plain-language interpretation, and access-description writing to help a general audience think about the object as a historical accessibility technology.

The central argument of the project is that **accessibility is part of cultural heritage**. Accessibility should not be understood only as a modern support feature added after a collection has already been interpreted. This object shows that access has a history. It reveals how religious knowledge, literacy, disability, and material design intersected in the nineteenth century. By focusing on the tactile qualities of the page, the project asks viewers to consider who books are made for, who is imagined as a reader, and what can be lost when tactile heritage is digitised only as flat images.

The project uses human–AI collaboration as part of its creative and interpretive process. Generative AI is used to draft alternative labels, access-description text, and explanatory wording. Python is used to create visual aids that make surface detail easier to notice in a digital environment. However, the AI is not treated as an authority. The human role remains responsible for selecting the object, choosing fragments, checking factual claims, editing language, identifying ethical risks, and deciding how the object should be presented to the public.

## Context

The selected object is a raised-type New Testament made for blind readers. It belongs to the University of Glasgow Special Collections and is identified as **Sp Coll RF 256 / RF 257**. As a religious book, it belongs to a long tradition of biblical printing and Christian reading practices. As a tactile book, however, it also belongs to the history of accessibility, disability, education, and reading technologies.

The Bible has had a major influence on Western language, culture, literature, art, and religious practice. For many communities, biblical texts have shaped moral teaching, collective memory, and personal belief. Yet access to biblical knowledge has never been equal. Historically, many people encountered scripture not through private reading, but through sermons, oral teaching, stained glass, paintings, public worship, and communal interpretation. For blind and visually impaired readers, access to printed scripture raised additional questions about format, independence, and material design.

The raised-type New Testament is therefore not only important because of what it says as a biblical text. It is important because of how it was made to be read. Its high-relief letters show that the page was designed for the hand as well as the eye. This changes the way the object can be interpreted. It is not only a religious artefact or a rare book. It is also evidence of an attempt to make reading accessible through material form.

This project approaches the object as a cultural heritage item connected to disability history. It does not present disability as an inspirational theme or as a problem to be solved. Instead, it focuses on design, access, and agency. The object helps open a wider conversation about how cultural institutions describe, digitise, and present materials that were made for different bodies and different reading practices.

In digital heritage, tactile objects can be difficult to represent. A photograph can show the page, but it cannot fully communicate height, pressure, resistance, or the movement of fingers across raised letters. This project therefore asks how a digital mini-exhibit can make tactile features more noticeable without pretending to replace the physical object. The answer developed here is a human–AI workflow that combines object photography, computational visual aids, AI-assisted public writing, and human ethical editing.

## How Generative AI Works in This Project

Generative AI refers to artificial intelligence systems that produce new content based on patterns learned from training data. In this project, generative AI is used mainly for text. It helps draft alternative exhibit labels, access descriptions, and explanatory wording.

The text-based AI model used in this project can be understood through the basic machine-learning workflow:

| Machine-learning stage | How it appears in this project |
|---|---|
| Training data | The model has previously been trained on large collections of text from many sources. The exact training data is not fully visible to the user. |
| Algorithm/model | The model uses a transformer-based architecture that predicts likely sequences of words based on input prompts and learned patterns. |
| New input | The human author provides prompts describing the object, audience, tone, and ethical constraints. |
| Output | The model generates draft labels, descriptions, or explanations. |
| Human verification | The human author checks the output for accuracy, clarity, tone, and ethical suitability. |

Transformer models are especially useful for text generation because they use attention mechanisms to consider relationships between words and phrases. This allows them to generate fluent text. However, fluency does not guarantee truth. A model may produce confident but unsupported claims, especially when asked about a specific archival object. For this reason, the AI output in this project is treated as a draft, not as evidence.

The project also uses Python image processing. This is not generative AI in the same sense as a text model. It does not invent a new image or reconstruct missing features. Instead, it applies computational transformations to existing photographs. For example, a photograph can be converted into grayscale, sharpened, or edge-enhanced to make surface patterns easier to notice. This supports the public-engagement aim, but the resulting image must be labelled as a visual aid rather than as a scientific measurement of texture.

The project therefore distinguishes between:

- **archival evidence**: the original object and photographs;
- **computational visual aids**: processed images that help viewers notice features;
- **AI-assisted interpretation**: draft text generated by AI and edited by the human author;
- **human judgement**: the final decision-making process that checks meaning, tone, ethics, and accuracy.

```python
# Code Cell 4: Build a contact sheet as the visual entry point to the exhibit.
# The contact sheet is generated from the repository images so that the notebook remains reproducible.

thumbs = []

for path in image_paths:
    im = Image.open(path).convert("RGB")
    im.thumbnail((220, 220))
    thumbs.append((path.name, im.copy()))

cols = 2
cell_w, cell_h = 520, 285
rows = int(np.ceil(len(thumbs) / cols))

sheet = Image.new("RGB", (cols * cell_w, rows * cell_h), "white")
draw = ImageDraw.Draw(sheet)

for idx, (name, im) in enumerate(thumbs):
    x = (idx % cols) * cell_w + 20
    y = (idx // cols) * cell_h + 20
    sheet.paste(im, (x, y))
    draw.text((x + 240, y + 70), name, fill="black")

contact_path = CONTACT_DIR / "selected_images_contact_sheet.jpg"
sheet.save(contact_path, quality=90)

display(IPImage(filename=str(contact_path)))
```

### Step 5: Creating relief-enhanced visual aids

The next step uses Python to create visual aids from the photographs. These outputs are called **relief-enhanced visual aids** rather than scans or measurements. This wording matters because the code does not scientifically measure the height of the raised type. It only transforms the image to make contrast and surface variation easier to notice.

This is useful for sighted online viewers who cannot handle the object directly. The processed image helps them notice features that may be less obvious in an ordinary photograph.

At the same time, the project is careful about the limits of this method. A relief-enhanced image cannot reproduce touch. It cannot show pressure, resistance, scale, or the movement of fingertips across the page. It is an interpretive aid, not a replacement for the object.

```python
# Code Cell 6: Functions for cropping and creating relief-enhanced visual aids.

def crop_by_percentage(image, crop_pct):
    """
    Crop an image using percentage coordinates.
    crop_pct should be (left, top, right, bottom), with each value between 0 and 1.
    """
    width, height = image.size
    left = int(width * crop_pct[0])
    top = int(height * crop_pct[1])
    right = int(width * crop_pct[2])
    bottom = int(height * crop_pct[3])
    return image.crop((left, top, right, bottom))


def create_relief_visual_aid(image):
    """
    Create an interpretive visual aid that makes surface variation easier to notice.
    
    This does not measure physical relief. It enhances contrast and edge information
    in the photograph so that sighted online viewers can better notice surface detail.
    """
    # Convert to grayscale so that tonal differences are easier to process.
    gray = ImageOps.grayscale(image)

    # Increase contrast to make surface variation more noticeable.
    contrast = ImageEnhance.Contrast(gray).enhance(1.8)

    # Sharpen image details.
    sharp = ImageEnhance.Sharpness(contrast).enhance(2.0)

    # Find edges and blend them with the enhanced grayscale image.
    edges = sharp.filter(ImageFilter.FIND_EDGES)
    edges = ImageEnhance.Contrast(edges).enhance(1.7)

    blended = Image.blend(sharp.convert("RGB"), edges.convert("RGB"), 0.35)

    return blended


print("Cropping and relief-enhancement functions are ready.")
```

The relief-enhanced image is not a scientific scan of the page. It does not measure the height of the raised letters. It is an interpretive visual aid that makes surface variation easier to notice in a flat digital image.

This distinction matters because the project should not overclaim what Python image processing can do. The processed images support public interpretation, but they do not replace the physical object or the tactile experience of reading by touch.

```python
# Code Cell 7: Generate cropped images and relief-enhanced visual aids.

fragment_plan = [
    {
        "fragment": "Fragment 1",
        "source": choose_image(0),
        "crop_pct": (0.05, 0.05, 0.95, 0.75)
    },
    {
        "fragment": "Fragment 2",
        "source": choose_image(1),
        "crop_pct": (0.10, 0.10, 0.90, 0.80)
    },
    {
        "fragment": "Fragment 3",
        "source": choose_image(2),
        "crop_pct": (0.05, 0.15, 0.95, 0.90)
    }
]

fragment_records = []

for item in fragment_plan:
    source_path = item["source"]
    original = Image.open(source_path).convert("RGB")

    cropped = crop_by_percentage(original, item["crop_pct"])
    cropped.thumbnail((900, 900))

    relief = create_relief_visual_aid(cropped)

    fragment_id = item["fragment"].lower().replace(" ", "_")

    crop_path = PROCESSED_DIR / f"{fragment_id}_crop.jpg"
    relief_path = PROCESSED_DIR / f"{fragment_id}_relief_visual_aid.jpg"

    cropped.save(crop_path, quality=90)
    relief.save(relief_path, quality=90)

    fragment_records.append({
        "fragment": item["fragment"],
        "source image": source_path.name,
        "crop path": str(crop_path.relative_to(PROJECT_ROOT)),
        "relief visual aid path": str(relief_path.relative_to(PROJECT_ROOT))
    })

processed_fragments = pd.DataFrame(fragment_records)
display(processed_fragments)
```

The table above shows the generated image outputs. These images are used directly in the mini-exhibit. The cropped image provides a closer photographic view, while the relief-enhanced visual aid helps the audience notice surface variation.

The purpose is not to make the page look more dramatic. The purpose is to support the visitor’s understanding of raised type as an access technology.

### Step 7: A small technical summary

The following code gives a simple technical summary of the processed fragments. It calculates the average brightness and a rough edge-density value for each crop and relief-enhanced visual aid.

This is not a scientific measurement of raised type. It is included to show that the code is doing a visible transformation to the image data. The values help explain the difference between the photographic crop and the enhanced visual aid in a way that is transparent to the reader.

```python
# Code Cell 8: Create side-by-side comparison images.

def make_comparison_image(crop_path, relief_path, title):
    """
    Create a side-by-side comparison image for the mini-exhibit.
    """
    crop = Image.open(crop_path).convert("RGB")
    relief = Image.open(relief_path).convert("RGB")

    target_height = 420

    def resize_to_height(im, height):
        ratio = height / im.height
        return im.resize((int(im.width * ratio), height))

    crop = resize_to_height(crop, target_height)
    relief = resize_to_height(relief, target_height)

    padding = 30
    title_h = 60
    label_h = 40

    width = crop.width + relief.width + padding * 3
    height = target_height + title_h + label_h + padding

    canvas = Image.new("RGB", (width, height), "white")
    draw = ImageDraw.Draw(canvas)

    draw.text((padding, 20), title, fill="black")
    draw.text((padding, title_h), "Photographic crop", fill="black")
    draw.text((crop.width + padding * 2, title_h), "Relief-enhanced visual aid", fill="black")

    canvas.paste(crop, (padding, title_h + label_h))
    canvas.paste(relief, (crop.width + padding * 2, title_h + label_h))

    return canvas


comparison_paths = []

for _, row in processed_fragments.iterrows():
    crop_path = PROJECT_ROOT / row["crop path"]
    relief_path = PROJECT_ROOT / row["relief visual aid path"]

    comparison = make_comparison_image(
        crop_path,
        relief_path,
        f"{row['fragment']}: photograph and interpretive visual aid"
    )

    comparison_path = PROCESSED_DIR / f"{row['fragment'].lower().replace(' ', '_')}_comparison.jpg"
    comparison.save(comparison_path, quality=90)
    comparison_paths.append(comparison_path)

    display(IPImage(filename=str(comparison_path)))
```

The comparison images are the technical centre of the project. They show how a photograph can be transformed to support public interpretation.

The processed view may make surface detail easier to notice, but it should not be confused with the physical surface of the page. This is important because the object’s meaning depends on touch, and touch cannot be fully reproduced in a notebook.

```python
# Code Cell 9: Create a small technical summary.
# This is not a scientific measurement of raised type.
# It only describes simple visual differences between original crops and processed visual aids.

def image_summary(path):
    """
    Return simple brightness and edge-density values for an image.
    These values are used only as a basic technical description.
    """
    im = Image.open(path).convert("L").resize((400, 400))
    arr = np.asarray(im).astype(float)

    brightness = arr.mean()

    horizontal_change = np.abs(np.diff(arr, axis=1)).mean()
    vertical_change = np.abs(np.diff(arr, axis=0)).mean()
    edge_density = horizontal_change + vertical_change

    return brightness, edge_density


summary_rows = []

for _, row in processed_fragments.iterrows():
    crop_path = PROJECT_ROOT / row["crop path"]
    relief_path = PROJECT_ROOT / row["relief visual aid path"]

    crop_brightness, crop_edges = image_summary(crop_path)
    relief_brightness, relief_edges = image_summary(relief_path)

    summary_rows.append({
        "fragment": row["fragment"],
        "crop average brightness": round(crop_brightness, 2),
        "crop edge-density estimate": round(crop_edges, 2),
        "relief aid average brightness": round(relief_brightness, 2),
        "relief aid edge-density estimate": round(relief_edges, 2)
    })

technical_summary = pd.DataFrame(summary_rows)
display(technical_summary)
```

The edge-density estimate is included carefully. It does not prove where the raised letters are or how high they are. It simply shows that the processed images contain stronger local contrast than the original crops.

For the audience, this supports the visual task of noticing surface variation. For the project methodology, it shows that the visual aids are generated through a transparent computational process rather than manually edited without explanation.

## Discussion: Access, AI, and Public Interpretation

The project shows that human–AI collaboration can support cultural heritage interpretation when it is used with clear limits. Generative AI is helpful for drafting alternative public labels and access descriptions because it can quickly generate different phrasings. This is useful when the aim is to communicate with non-specialist audiences. However, the AI cannot decide what is historically accurate, ethically appropriate, or respectful.

This limitation is especially important in a project about disability and access. AI-generated language can easily become too general or too sentimental. It may describe an accessibility object through ideas of inspiration or overcoming, rather than through design, rights, and material practice. Human editing is therefore not a minor final correction. It is a central part of the methodology.

The relief-enhanced visual aids also require careful interpretation. They make the raised surface easier to notice, but they do not reproduce touch. A viewer may be able to see more contrast in the processed image, but that does not mean they understand the bodily experience of reading with fingertips. This is why the notebook repeatedly labels the processed images as interpretive visual aids rather than scientific scans.

The project therefore treats access in two ways. First, access is the subject of the exhibit: the object itself is a historical example of designing reading for blind readers. Second, access is the method of the exhibit: the notebook uses plain language, image descriptions, tables, and visual comparisons to make the interpretation easier to follow.

This matters because cultural heritage is not automatically accessible just because it has been digitised. A digital image may increase visibility, but it can still exclude audiences if it lacks description, context, permissions clarity, or accessible design. For tactile objects, digitisation can create a particular problem: it shares the object visually while losing the tactile quality that made the object significant.

The project also shows that AI should not be understood as replacing human expertise. In this portfolio, AI supports drafting and variation, while Python supports visual transformation. The human author remains responsible for the curatorial argument, ethical judgement, factual checking, accessibility decisions, and final presentation. This is the most important part of the collaboration.

## Ethical, Legal, and Archival Considerations

The project raises several ethical and legal considerations.

First, the object images are photographs of a University of Glasgow Special Collections object. They are used here for coursework purposes. The object should be credited as:

**Tactile New Testament printed in high relief for blind readers, c. 1850, University of Glasgow Special Collections, Sp Coll RF 256 / RF 257.**

The images should not be treated as openly licensed unless University of Glasgow Special Collections confirms the relevant reuse terms. Before sharing the images beyond coursework assessment, permissions and licensing conditions should be checked.

Second, the project must be clear about the difference between evidence and interpretation. The photographs show visible features of the object. The relief-enhanced images are computational interpretations of those photographs. The labels and access descriptions are public-facing interpretations written by the human author with AI assistance. These categories should not be confused.

Third, the project works with disability-related heritage. This requires careful language. The notebook avoids treating blind readers as objects of pity or as inspirational symbols. It focuses instead on access, design, material reading practices, and the social conditions that shape who can read independently.

Fourth, the project uses generative AI. AI-generated text may contain unsupported claims or unsuitable tone. For that reason, the AI is used only as a drafting tool. The human author verifies and revises the outputs before they appear in the final exhibit.

Finally, the project acknowledges the limits of digital access. A notebook can help audiences understand the object, but it cannot replace archival handling, expert cataloguing, or tactile experience. This limitation is not hidden. It is part of the project’s argument.
